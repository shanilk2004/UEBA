```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import pickle

traindf=pd.read_parquet('UNSW_NB15_training-set.parquet')

print(f"data uplaoded,total rows={len(traindf)}")

mvp_features=['dur','sbytes','dbytes','sload','dload']

X_train=traindf[mvp_features].dropna()

print(f"now the data is cleaned, remain rows={len(X_train)}")

bs_model=IsolationForest(contamination=0.01,random_state=42)
bs_model.fit(X_train)


with open('baseline_isolation_forest.pkl', 'wb') as f:
    pickle.dump(bs_model, f)
print("🧠 'baseline_isolation_forest.pkl' is ready for download!")

```
